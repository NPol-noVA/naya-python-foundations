```python
import pandas as pd

rf = pd.read_csv(
    r'C:\Users\USER\Downloads\nutr.csv',
    low_memory=False
)

print("Shape:")
print(rf.shape)

print("\nColumn Names:")
print(rf.columns)

print("\nData Types:")
print(rf.dtypes)

print("\nFirst 5 Rows:")
print(rf.head())
rf = rf.dropna(subset=["Low_Confidence_Limit"])

print("Shape after cleaning:")
print(rf.shape)
avg = rf["Data_Value"].mean()

print("Average Data_Value:")
print(avg)
location_counts = rf["LocationDesc"].value_counts()

print(location_counts)
location_avg = rf.groupby("LocationDesc")["Data_Value"].mean()

print(location_avg.sort_values(ascending=False).head())
```
